```python
import cv2
import numpy as np
import os
from datetime import datetime
```

```python
def markAttendance(name):
    with open('Attendance.csv', 'r+') as f:
        myDataList = f.readlines()
        nameList = []
        for line in myDataList:
            entry = line.split(',')
            nameList.append(entry[0])
        if name not in nameList:
            time_now = datetime.now()
            tString = time_now.strftime('%H:%M:%S')
            dString = time_now.strftime('%d/%m/%Y')
            f.writelines(f'\n{name},{tString},{dString}')
```

```python
def name_map(x):
    if x=='Messi':
        return 0
    if x=='Obama':
        return 1
    if x=='Sachin':
        return 2
   
```

```python
def get_name(x):
    if x==0:
        return 'Messi'
    if x==1:
        return 'Obama'
    if x==2:
        return 'Sachin'
```

```python
def detect_face(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face_cascade =cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces =face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))

    if(len(faces)==0):
        return None,None # if there is no face to detect
    (x,y,w,h) = faces[0] # first element in faces[]
    
    return gray[y:y+w,x:x+h] ,faces[0] # to crop the image
```

```python
for dir_name in dirs:

    subject_dir_path ='data' + '/' + dir_name    
    subject_img_names=os.listdir(subject_dir_path)

    for img_names in subject_img_names:
        img_path=subject_dir_path+'/' +img_names
        print(img_path)
```

```python
def prepare_training_data(data_folder_path):
    dirs=os.listdir(data_folder_path)

    faces= []
    labels= []
    
    for dir_name in dirs:

        label=dir_name

        subject_dir_path = data_folder_path+'/' +dir_name
        subject_img_names =os.listdir(subject_dir_path)

        for img_name in subject_img_names:
            img_path = subject_dir_path+'/'+img_name

            image=cv2.imread(img_path)
            cv2.imshow('prepare_training_data...',image)
            cv2.waitKey(1000)
                

            face,rect=detect_face(image)

            if face is not None:
                faces.append(face)
                labels.append(name_map(label))
                print(labels)
    cv2.waitKey(1)
    cv2.destroyAllWindows()

    return faces ,labels
            
```

```python
print('preparing data')

faces , labels=prepare_training_data('data')

print('data prepared')
```

```python
print('total faces',len(faces))
```

```python
print('total labels',len(labels))
```

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
```

```python
face_recognizer.train(faces,np.array(labels))
```

```python
def draw_rectangle(img,rect):

    (x,y,w,h)=rect
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
```

```python
def draw_text(img,text,x,y):
    cv2.putText(img,text,(x,y),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,1,(0,0,255),2)
    markAttendance(text)
```

```python


def predict(test_img):
    img = test_img.copy()
    
    face,rect=detect_face(img)
    x,y= rect[0],rect[1]

    label = face_recognizer.predict(face)
    
    label_text = get_name(label[0])
    
    draw_rectangle(img,rect)

    draw_text(img,label_text,rect[0],rect[1])

    return img
```
